### Handling missing data

```python
a = pd.Series(np.random.randint(0, 10, 4), index=list("ABCD"))
b = pd.Series(np.random.randint(0, 10, 4), index=list("ABCE"))
print("a =\n", a)
print("b =\n", b)
c = a+b
print("c =\n", c)
```

### Fill the missing values with 0 

```python
# fills the missing indices with 0 and then add the series
a.add(b, fill_value=0)
```

```python
mean = a.mean()
print("Mean of series a =", mean)
a.add(b, fill_value=mean)
```

In case of DataFrame, we need to stack the dataframe 'A' in order to calculate its mean

```python
A = pd.DataFrame(np.random.randint(0, 10, [2, 3]), index=list("XY"))
print("A =", "\n", A)
B = pd.DataFrame(np.random.randint(0, 10, [2, 3]), index=list("XZ"))
print("B =", "\n", B)
```

```python
C = A + B
C
```

### Filling the missing values with Mean value

```python
meanVal = A.stack().mean()
print("Mean of A =", meanVal)
A.add(B, fill_value=meanVal)
```

axis=0 means aligning index of Series A.loc['X'] with **index of DataFrame A** \
Index of A = [X, Y] \
Index of A.loc['X'] = [0, 1, 2] \
Because of no overlap the above result is calculated

```python
A.subtract(A.loc['X'], axis=1)
```

#### Merge

```python
A = pd.DataFrame([["JAKE", 2, 4], ["RACE", 9, 0]], index=list("XY"), columns=list("DBC"))
B = pd.DataFrame([["JAKE", 23, 44], ["RACE", 29, 10]], index=list("XY"), columns=list("XEF"))
display("A", A); display("B", B)
```

```python
# Merging on columns
pd.merge(A, B, left_on="D", right_on="X")
```

```python
# Merging on Index
pd.merge(A, B, left_index=True, right_index=True)
```

```python
# Join (same as merging on index)
A.join(B)
```

### Task: Merge the following 3 datasets and rank US states and territories by their 2010 population density

```python
# !curl -O https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-population.csv
# !curl -O https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-areas.csv
# !curl -O https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-abbrevs.csv
```

```python
pop = pd.read_csv("state-population.csv")
areas = pd.read_csv("state-areas.csv")
abbrevs = pd.read_csv("state-abbrevs.csv")
```

```python
display(pop.head())
display(areas.head())
display(abbrevs.head())
```

#### Finding missing values (NaN) in all datasets

#### Merging Population and Abbreviation dataset

```python
popAbbrMerged = pd.merge(pop, abbrevs, how="outer", left_on="state/region", right_on="abbreviation")
popAbbrMerged.head()
```

Checking if merged dataset has mismatches (NaN)

```python
popAbbrMerged.isnull().any()
```

```python
popAbbrMerged.loc[popAbbrMerged["state"].isnull()]
```

```python
popAbbrMerged.loc[popAbbrMerged["state"].isnull(), "state/region"].unique()
```

The columns "state" and "abbreviation" of table "Abbreviation" did not have "Puerto Rico" and "United States", and "PR" and "USA" respectively, which is the cause of NaN values. \
This can be fixed by filling in appropriate entries

```python
popAbbrMerged.loc[popAbbrMerged["state/region"] == "PR", "state"] = "Puerto Rico"
popAbbrMerged.loc[popAbbrMerged["state/region"] == "USA", "state"] = "United States"
popAbbrMerged.loc[popAbbrMerged["state/region"] == "PR", "abbreviation"] = "PR"
popAbbrMerged.loc[popAbbrMerged["state/region"] == "USA", "abbreviation"] = "USA"
```

```python
# Checking for NaN values again (ignore the population NaN values as discussed earlier)
popAbbrMerged.isnull().any()
```

#### Merging the above merged dataset with Areas dataset

```python
finalDf = pd.merge(popAbbrMerged, areas, how="outer", left_on="state", right_on="state")
display("(Rows, Columns) = " + str(finalDf.shape))
finalDf.head()
```

Checking for NaN

```python
finalDf.isnull().any()
```

```python
# Finding states that have no area values
finalDf.loc[finalDf["area (sq. mi)"].isnull(), "state"].unique()
```

**Since we need to rank states and regions of USA, we don't need rows that contain data of USA as whole. Therefore those rows are dropped**

```python
display("Before dropping USA NaN values, (Rows, Columns) = " + str(finalDf.shape))
indicesOfUSARows = finalDf.loc[finalDf["state/region"] == "USA"].index
finalDf.drop(indicesOfUSARows, inplace=True)
display("After dropping USA NaN values, (Rows, Columns) = " + str(finalDf.shape))
```

```python
finalDf
```

```python
finalDf["density"] = finalDf["population"] / finalDf["area (sq. mi)"]
finalDf.head()
```

```python
finalDf.isnull().any()
```

Density has  NaN values, because some rows had NaN population values right from the very beginning, so drop those values as well.

```python
finalDf.dropna(inplace=True)
display("(Rows, Columns) = " + str(finalDf.shape))
display(finalDf.head())
display(finalDf.isnull().any())
```

```python
year2010Df = finalDf.loc[(finalDf["year"] == 2010) & (finalDf["ages"] == "total")].copy()
year2010Df.set_index("state/region", inplace=True)
year2010Df.sort_values(by="density", ascending=False, inplace=True)
display(year2010Df[["state", "density"]].head())
display(year2010Df[["state", "density"]].tail())
```

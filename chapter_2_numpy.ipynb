## copy()

If changes are made to a subarray/submatrix, then those changes are reflected in the original matrix as well.
To avoid this and make a deep copy, use copy() function

```python
mat = np.random.randint(0, 10, [3, 2])
print("Original mat:\n", mat)
shallowCopy = mat[:2,:]
deepCopy = mat[:2,:].copy()
print("Shallow copy:\n", shallowCopy)
print("Deep copy:\n", deepCopy)
shallowCopy[0, 0] = 100
deepCopy[1, 1] = 500
print("Shallow copy affected:\n", shallowCopy)
print("Deep copy affected:\n", deepCopy)
print("Original mat affected:\n", mat)
```
